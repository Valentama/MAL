```python
from bs4 import BeautifulSoup
import pandas as pd

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
#install browser
chromedriver_autoinstaller.install()

#menggunakan browser yang telah di install
browser = webdriver.Chrome()
```

```python
AJudul = []
ATEps = []
ADate = []
AFav = []
ARating = []
for index in range(0, 1001, 50):
    page = browser.get(f'https://myanimelist.net/topanime.php?type=favorite&limit={index}')
    soup = BeautifulSoup(browser.page_source, 'lxml')
    
    Judul = soup.find_all('h3', class_='hoverinfo_trigger')
    for J in Judul:
        AJudul.append(J.text.strip())
        
    TEps = soup.find_all('div', class_='information')
    for T in TEps:
        ATEps.append((T.text.strip().splitlines()[0]))
        
    Date = soup.find_all('div', class_='information')
    for D in Date:
        ADate.append((D.text.strip().splitlines()[1]))
        
    Fav = soup.find_all('div', class_='information')
    for F in TEps:
        AFav.append((F.text.strip().splitlines()[2]))
        
    Rating = soup.find_all('td', class_='score ac fs14')
    for R in Rating:
        ARating.append(R.text.strip())
```

```python
Data = {
    'Judul' : AJudul,
    'Type & Eps' : ATEps,
    'Date' : ADate,
    'Fav' : AFav,
    'Rating' : ARating,
}
pd.DataFrame(Data).to_excel('MAL Top Anime List.xlsx')
```
